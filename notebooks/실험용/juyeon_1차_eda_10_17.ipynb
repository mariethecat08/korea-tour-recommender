```python
# 문7 관련 컬럼 추출
X = df[partner_cols].fillna(0).astype(int)

# 0이 아닌 값은 모두 1로 바꾸기 -> 각 1,2,...,8로 저장되어 있음
X = X.applymap(lambda x: 1 if x != 0 else 0).astype(int)

# 각 행의 조합
combo_series = X.apply(lambda row: '+'.join( [col.split('(')[1].replace(')', '') for col, val in row.items() if val == 1]), axis=1)

# 조합 빈도수
combo_counts = combo_series.value_counts()
combo_counts.head(20)
```

```python
# import numpy as np
combo = combo_series.copy()

# 각 그룹 조건 설정
cond_혼자 = combo.str.contains('혼자', na=False)
cond_가족 = combo.str.contains('배우자|자녀|부모님|가족', na=False)
cond_친구 = combo.str.contains('친구', na=False)
cond_직장 = combo.str.contains('직장', na=False)
cond_기타 = combo.str.contains('기타', na=False)

# 그룹화 (복합형 먼저 처리해야 함)
conditions = [(cond_가족 & cond_친구),
               cond_가족,
               cond_친구,
               cond_혼자,
               cond_직장,
               cond_기타 ]
choices = ['복합형(가족+친구)', '가족형', '친구형', '혼자형', '직장형', '기타형']

df['여행동반자_그룹'] = np.select(conditions, choices, default='기타형')
df['여행동반자_그룹'].value_counts()
```

- 혼자형 : 혼자
- 가족형 : 배우자 / 자녀/ 부모님 / 친지
- 친구형 : 친구
- 직장형 : 직장동료
- 복합형 : 가족 + 친구  **2개이상의 관계 유형이 혼합**
- 기타형 : 기타

(내 생각에 장점)
- 그룹이 동일한 의미 단위(관계 유형)로 정의되어서 일관성있음
- 해석력이 좋다

# 문9-5. 주요 이용 숙박시설

```python
df['문9-5. 주요 이용 숙박시설(1. 호텔)'].value_counts()
```

```python
import numpy as np

# 숙박시설 관련 컬럼 자동 선택
stay_type_cols = [c for c in df.columns if c.startswith('문9-5. 주요 이용 숙박시설(')]

# 선택된 숙박시설명을 합치는 함수
def merge_stay_types(row):
    selected = [col.split('(')[-1].replace(')', '') for col in stay_type_cols if not pd.isna(row[col])]
    if len(selected) == 0:
        return np.nan
    elif len(selected) == 1:
        return selected[0]
    else:
        return '+'.join(selected)

# 통합 컬럼 생성
df['주요 이용 숙박시설_통합'] = df.apply(merge_stay_types, axis=1)

# 결과 확인
print(df['주요 이용 숙박시설_통합'].value_counts(dropna=False).head(20))
```

```python
# 통합된 숙박시설 문자열 컬럼
combo = df['주요 이용 숙박시설_통합'].copy()

# 키워드별 포함 여부 (True/False)
cond_호텔 = combo.str.contains('호텔', na=False)
cond_친척 = combo.str.contains('친척|친구', na=False)
cond_게하 = combo.str.contains('게스트하우스|호스텔', na=False)
cond_모텔 = combo.str.contains('모텔|여관', na=False)
cond_콘도 = combo.str.contains('콘도|리조트', na=False)
cond_민박 = combo.str.contains('민박|펜션', na=False)
cond_학교 = combo.str.contains('기숙사|연수원', na=False)
cond_기타 = combo.str.contains('기타', na=False)
cond_숙박X = combo.str.contains('숙박하지 않음', na=False)

# 각 행마다 포함된 유형 개수 세기
type_count = (
    cond_호텔.astype(int)
    + cond_친척.astype(int)
    + cond_게하.astype(int)
    + cond_모텔.astype(int)
    + cond_콘도.astype(int)
    + cond_민박.astype(int)
    + cond_학교.astype(int)
    + cond_기타.astype(int)
)

# 그룹화 로직
conditions = [
    type_count > 1,  # 두 개 이상 선택 시 복합형
    cond_호텔,
    cond_친척,
    cond_게하,
    cond_모텔,
    cond_콘도,
    cond_민박,
    cond_학교,
    cond_기타,
    cond_숙박X
]

choices = [
    '복합형',
    '호텔형',
    '친척/친구형',
    '게스트하우스형',
    '모텔형',
    '리조트형',
    '민박형',
    '기숙사형',
    '기타형',
    '숙박안함형'
]

# 최종 그룹 생성
df['숙박시설_그룹'] = np.select(conditions, choices, default='기타형')
print(df['숙박시설_그룹'].value_counts())
```
